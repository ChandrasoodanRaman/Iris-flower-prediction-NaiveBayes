```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
dir(iris)
```

```python
df=pd.DataFrame(data=iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target']=iris.target
```

```python
df.head()
```

```python
df=df.drop(df[df['target']==2].index)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(df,df['target'],test_size=0.2)
```

```python
df1=df[df['target']==0]
df2=df[df['target']==1]
df.describe()
```

```python
f1=df1.describe()
f2=df2.describe()
f1#sestoso(target=0)
f2#versicolor(target=1)
```

```python
m1=f1.iloc[1]#mean row for target=0(sestoso)
m2=f2.iloc[1]#mean row for target=1(versicolor)
```

```python
std1=f1.iloc[2]#std row for target=0(sestoso)
std2=f2.iloc[2]#std row for target=1(versicolor)
```

```python
X_test
```

```python
#a-input b-mean c-std dev
def gauss(a,b,c):
    d=-1*(((a-b)/c)**2)  #−((x − µ)/σ)^2 
    d=(math.exp(d))/c #e^−((x − µ)/σ)^2 /σ
    return d
```

```python
#iloc[0][4] 0th row=count  4th column=target
total=df.describe().iloc[0][4] #total count (number of samples) in the DataFrame df 
total1=df1.describe().iloc[0][4]#total count (number of samples) in the DataFrame df1(sestoso)
total2=df2.describe().iloc[0][4]#total count (number of samples) in the DataFrame df2(versicolor)

# predicted labels for the test data.
predclass=[]

for index,row in X_test.iterrows():
    #two probability variables, p1 and p2, to 1.
    p1=1
    p2=1
    
    #target=0
    for i in range(0,4):
        
        #p1=p1*probability density
        p1=p1*gauss(row[i],m1[i],std1[i])#gauss(input,mean,stddev)
    #p1 is multiplied by the prior probability of class 0, prior probability total1/total  no of sample in df1/total sample
    p1=p1*(total1/total)
    
    
    #target=1
    for i in range(0,4):
        p2=p2*gauss(row[i],m2[i],std2[i])
    p2=p2*(total2/total)
    
    #to determine which class has a higher probability
    if p1>p2:
        predclass.append(0)
    else:
        predclass.append(1)
        
```

```python
predclass
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,predclass)
```
